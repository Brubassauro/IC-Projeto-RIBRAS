```
import numpy as np
import matplotlib.pyplot as plt

def lista(A):
  a1=A.replace('\r\n','')
  a1=A.replace('\\r\\n','')
  a1=a1.split(',')
  a2=[]
  for i in range(len(a1)-2):
    a2+=[float(a1[1+i])]
  return a2

def graf(L,cor,tit):
  th=[]
  div=[]
  for i in range(int(len(L)/3)):
    th+=[L[3*i]]
    div+=[L[2+3*i]]
  plt.plot(th,div,color=cor,label=tit)
  
graf(lista(he6),'blue','6He')
plt.title('Potencial PSP para 6He em 9Be (16.1MeV)')
plt.xlabel('Theta lab [Â°]')
plt.ylabel('Sig/SigR')
plt.xlim(0,120)
plt.yscale('log')
#plt.legend()
plt.grid()

plt.show()
```
